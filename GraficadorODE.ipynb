```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = np.genfromtxt("datos.dat",dtype=None)
datosDx=np.zeros([len(datos)])
datosDy=np.zeros([len(datos)])
datosVx=np.zeros([len(datos)])
datosVy=np.zeros([len(datos)])
for i in range(len(datos)):
    listaDatos = ((datos[i]).decode('UTF8')).split(",")
    datosDx[i] = float(listaDatos[0])
    datosDy[i] = float(listaDatos[1])
    datosVx[i] = float(listaDatos[2])
    datosVy[i] = float(listaDatos[3])
```

```python
plt.figure()
plt.plot(datosDx,datosDy)
```
